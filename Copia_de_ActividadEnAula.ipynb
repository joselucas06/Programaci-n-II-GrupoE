```python
import pandas as pd
df=pd.read_csv('doctores.txt',sep='|',header=None)
df.columns=["Nombre","Especialidad","AñosExperiencia","Sexo","Hospital"]
display(df)
```

```python
print("Promedio de Experiencia: ",df["AñosExperiencia"].mean())
print("Maximo de Experiencia: ",df["AñosExperiencia"].max())
print("Minimo de Experiencia: ",df["AñosExperiencia"].min())
```

```python
#filtrar por especialidad
print("Doctores de cardiologia:")
print(df[df["Especialidad"]=="Cardiologia"])
print("Doctores de Neurologia:")
print(df[df["Especialidad"]=="Neurologia"])
print("Doctores de Pediatria:")
print(df[df["Especialidad"]=="Pediatria"])
print("Doctores de Traumatologia:")
print(df[df["Especialidad"]=="Traumatologia"])
print("Doctores de Medicina General:")
print(df[df["Especialidad"]=="Medicina General"])
#filtrar por sexo
print("Doctores Femeninos:")
print(df[df["Sexo"]=="Femenino"])
print("Doctores Masculinos:")
print(df[df["Sexo"]=="Masculino"])
```

```python
#Grafico de barras por especialidad
import matplotlib.pyplot as plt
df['Especialidad'].value_counts().plot(kind='bar')
plt.title('Doctores por Especialidad')
plt.xlabel('Especialidad')
plt.ylabel('Cantidad')
plt.show()
```

```python
#Grafico de pastel por sexo
df['Sexo'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Doctores por Sexo')
plt.ylabel('')
plt.show()
```
